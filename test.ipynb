```python
import numpy as np
import matplotlib.ticker
import matplotlib.pyplot as plt
from math import pi
from src.lib.fluid import *
```

```python
nz = 20
Q = 1000
x = np.linspace(0, 1, nz)
qz = [Q*x, Q*x**2, Q*np.sin(2*pi*x)]
dh = 0.01
area = dh**2*pi/4
plenum = Plenum(1., 3, dh, area)
plenum.setMeanProperties(25, 1.)
plenum.setProperties(0.5, 1.)
plenum.setInletDensity()
```

```python
m0 = [0.15, 0.7, 0.15]
dp0 = 0
```

```python
sol = plenum.solveSystem(qz, dp0, m0, 100, 1e-6)
m = sol[1:]
```

```python
# Compute outlet density
rho_out = plenum.computeOutletDensity(qz, m)
print(rho_out)
```

```python
T_out = plenum.computeOutletTemperature(qz, m)
print(T_out)
```

```python
T_profile = plenum.computeTemperatureProfile(qz, m)
rho_profile = plenum.computeDensityProfile(qz, m)
# plot profiles in a same figure
fig, ax = plt.subplots(1, dpi=300)
for i in range(3):
    ax.plot(x, T_profile[i], label=f'T{i+1}(z)')
# twince axis
ax2 = ax.twinx()
for i in range(3):
    ax2.plot(x, rho_profile[i], label=f'ρ{i+1}(z)', linestyle='--')
ax.legend()
ax2.legend()
#ax.yaxis.set_major_formatter(matplotlib.ticker.FormatStrFormatter('%.2f'))
plt.show()
```
